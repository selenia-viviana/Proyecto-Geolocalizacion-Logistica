```python
from statistics import median
department_counts = {}
dis_horas_centro_min = 10
dis_horas_centro_inter = 18
dis_horas_centro_max = 30 
lst =[]

for i in range(0, 23):
    row_str = df_horas_bus.iloc[i]
    department = row_str.name
    row_int = [int(item) if item.isdigit() else item for item in row_str]
    row_int_filtered = [int(item) for item in row_int if isinstance(item, int) or (isinstance(item, str) and item.isdigit())]
    if row_int_filtered:
        min_value = __builtins__.min(row_int_filtered)
        max_value = __builtins__.max(row_int_filtered)
        sortedvalue = sorted(row_int_filtered)
        second_min_value = sortedvalue[1]
        median_value = median (row_int_filtered)
        count_greater_than_min = sum(1 for item in row_int if isinstance(item, int) and item < dis_horas_centro_min)
        count_greater_than_inter = sum(1 for item in row_int if isinstance(item, int) and item > dis_horas_centro_min and item <dis_horas_centro_inter)
        count_greater_than_max = sum(1 for item in row_int if isinstance(item, int) and item > dis_horas_centro_inter and item <dis_horas_centro_max)
        new_row = [min_value, second_min_value, max_value, median_value, count_greater_than_min,count_greater_than_inter,count_greater_than_max]
        department_counts[department] = [ median_value,min_value, max_value, count_greater_than_min,count_greater_than_inter,count_greater_than_max, second_min_value]
        lst.append(new_row)
sorted_departments = sorted(department_counts.items(), key=lambda item: item[0][2], reverse=True)
department_data = []
for department, counts in sorted_departments:
    department_data.append({
        'Department': department,
        'Min Value': counts[1],
        'Second Min Value': counts[6],
        'Max Value': counts[2],
        'Median Value': counts[0],
        'Count < 10hrs': counts[3],
        'Count < 18hrs': counts[4],
        'Count < 30hrs': counts[5]
    })
    
df = pd.DataFrame(department_data)
sorted_df = df.sort_values(by='Median Value')
sorted_df
   
```

```python
def find_divisors(n):
    divisors = []
    for i in range(2, n):  # Start from 2 and go up to n-1
        if n % i == 0:
            divisors.append(i)
    return divisors

# Example usage
number = 2
divisors = find_divisors(number)
divisors
```

```python
def find_divisors_with_condition(n):
    divisors = [i for i in range(2, n) if n % i == 0]

    # Check if there are exactly two divisors
    if len(divisors) == 1:
        # Return the divisors and the greater one
        return n
    else:
        return divisors, None

# Example usage
number = 2
divisors, greatest_divisor = find_divisors_with_condition(number)

if greatest_divisor:
    print(f"The divisors of {number} are {divisors}, and the greatest divisor is {greatest_divisor}.")
else:
    print(f"The divisors of {number} are {divisors}.")
```
